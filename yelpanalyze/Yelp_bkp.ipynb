```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
yelp = pd.read_csv('C:/Users/rajes/Desktop/NYDS/Web Scraping in Python/yelpanalyze/restaurants3.csv')
```

```python
yelp = yelp[['Name', 'Category', 'NumberReviews', 'Overallrating','PriceRange', 'Delivery']]
yelp
```

```python
#List of restaurants with higher number of reviews
yelp2 = yelp.sort_values(by='NumberReviews', ascending = False)
len_of_df = len(yelp2)
len_of_df
```

```python
#Considering only Top 25% of restaurants with highest number of reviews
percentage = 25
len_top_25_percent_df = int(len_of_df*percentage/100)
len_top_25_percent_df
```

```python
top_25_percent_df = yelp2[:len_top_25_percent_df]
top_25_percent_df
```

```python
#List of restaurants with higher ratings & higher reviews[Reviews > 1000]
ratings = yelp.sort_values(by=['Overallrating', 'NumberReviews'], ascending = False)
ratings2 = ratings.loc[ratings['NumberReviews'] >= 1000]
ratings2
```

```python
#Check if we have any null value 
yelp.isna().any()
```

```python
# Replacing N/a with NaN rather than deleting the records
yelp['Delivery'] = yelp['Delivery'].fillna('NaN')
yelp.isna().any()
```

```python
#Average of ratings with Delievry & without Delivery:
ratings2 = pd.DataFrame(yelp.groupby(['Delivery']).Overallrating.mean()).reset_index()
ratings2
```

```python
#Creating a plot to check if the delivery has impact on the Restaurant Business
sns.catplot(x="Delivery", y="Overallrating", kind ='box' ,data=yelp)
```

```python
# Replace the dollar value  with price range
yelp.PriceRange = yelp.PriceRange.replace({"$": "Under $10","$$": "$11-30","$$$":"$31-60","$$$$":"Above $61"})
yelp.PriceRange.value_counts()
```

```python
# Box plot between PriceRange and OverallRating 
sns.catplot(x="PriceRange", y="Overallrating", kind ='box' ,data=yelp);
```

```python
#  Identify  Top 20 Restaurants.
Top20 = (yelp.groupby('Name',  as_index=False)
       .agg({'NumberReviews':'sum', 'Overallrating':'mean'}))
Top20= Top20.sort_values(['NumberReviews'], ascending=[False])
Top20.head(20)
```
